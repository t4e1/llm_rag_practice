1. LangGraph란?
**LangGraph**는 대규모 언어 모델(LLM)의 워크플로우를 **그래프 기반**으로 설계하고 실행할 수 있게 해주는 프레임워크입니다.

### 왜 LangGraph가 필요할까?

전통적인 LLM 애플리케이션은 단순한 일방향 흐름으로 구성되어 있습니다:

```
입력 → LLM → 출력
```

하지만 실제 복잡한 작업에서는 다음과 같은 요구사항이 있습니다:
- **조건부 분기**: 상황에 따라 다른 경로로 진행
- **반복 처리**: 만족스러운 결과가 나올 때까지 재시도
- **다중 도구 사용**: 여러 외부 도구나 API 조합
- **상태 관리**: 과정 중 정보를 기억하고 활용

### LangGraph의 장점

1. **시각적 표현**: 복잡한 워크플로우를 그래프로 시각화
2. **유연한 제어**: 조건부 분기와 순환 처리 지원
3. **상태 관리**: 전체 과정에서 정보 공유 및 관리
4. **모듈화**: 재사용 가능한 컴포넌트로 구성
5. **디버깅**: 각 단계별 상태 추적 가능

### LangGraph vs 기존 방식

| 구분 | 기존 방식 | LangGraph |
|------|-----------|-----------|
| 구조 | 선형적 | 그래프 기반 |
| 분기 처리 | 제한적 | 유연한 조건부 분기 |
| 상태 관리 | 어려움 | 내장된 상태 관리 |
| 재사용성 | 낮음 | 높음 (모듈화) |
| 디버깅 | 어려움 | 단계별 추적 가능 |


## 2. LangGraph의 핵심 구성요소

LangGraph는 세 가지 핵심 구성요소로 이루어져 있습니다:

### 1. Node (노드)
- **역할**: 그래프에서 특정 작업을 수행하는 개별 단위
- **기능**: LLM 호출, 함수 실행, 도구 사용, 데이터 처리 등
- **특징**: 독립적이고 재사용 가능한 작업 단위

### 2. Edge (엣지)
- **역할**: 노드 간의 연결을 정의
- **기능**: 작업의 흐름과 조건부 분기 설정
- **특징**: 워크플로우의 경로와 조건을 결정

### 3. State (상태)
- **역할**: 그래프 전체에서 공유되는 데이터
- **기능**: 노드 간 정보 전달 및 흐름 제어
- **특징**: 전체 과정의 컨텍스트를 유지
